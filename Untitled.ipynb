## Encoding categorical variables

For columns with 2 unique values we will use label encoding, whereas for features with more than 2 variables we will implement one-hot encoding.

```python
# Create label encoder object
le = LabelEncoder()

encode_count = 0

# Encode columns
for col in app_train:
    # Select categorical columns with 2 or less unique values
    if app_train[col].dtype == 'object':
        if len(app_train[col].unique().tolist()) < 3:
        # Fit and transform training and test data
            app_train[col] = le.fit_transform(app_train[col])
            app_test[col] = le.fit_transform(app_test[col])
        
            encode_count += 1

print('{} columns were successfuly label encoded'.format(encode_count))        
```

```python
print('Training columns before one-hot encoding:', app_train.shape[1])
print('Test columns before one-hot encoding:', app_test.shape[1])

# One-hot encoding
app_train = pd.get_dummies(app_train)
app_test = pd.get_dummies(app_test)

print('\nTraining columns after one-hot encoding:', app_train.shape[1])
print('Test columns after one-hot encoding:', app_test.shape[1])
```

## Aligning training and test data

Due to one-hot encoding and the different unique values in the datasets, the number of  new features do not correspond. 
In order for our model to work on both training and test data, we need both of the datasets to have the same columns. 

```python
train_labels = app_train['TARGET']

# Align the data between the datasets using inner join
app_train, app_test = app_train.align(app_test, join='inner', axis=1)

# Add training labels app_train
app_train['TARGET'] = train_labels

print('Training set dimensions:', app_train.shape)
print('Test set dimensions:', app_test.shape)
```

# Feature engineering

### Polynomial features

```python
# Select dataframe for polynomial features
poly_features_train = app_train[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH']]
poly_features_test = app_test[['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH']]
poly_target = app_train['TARGET']

# Handle missing value (imputation)
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan, strategy='median')

poly_features_train = imputer.fit_transform(poly_features_train)
poly_features_test = imputer.transform(poly_features_test)

# Transform features
from sklearn.preprocessing import PolynomialFeatures

poly = PolynomialFeatures(degree=3)

poly.fit(poly_features_train)

poly_features_train = poly.transform(poly_features_train)
poly_features_test = poly.transform(poly_features_test)
```

```python
print("New features created:", str(poly_features_train.shape[1] - 4), '\n')

poly_feat_names = poly.get_feature_names(input_features=['EXT_SOURCE_1', 'EXT_SOURCE_2', 'EXT_SOURCE_3', 'DAYS_BIRTH'])

print('Few polynomial features:', poly_feat_names[-10:])
```

```python
# Convert features to Dataframe
poly_features_train = pd.DataFrame(poly_features_train, columns=poly_feat_names)
poly_features_test = pd.DataFrame(poly_features_test, columns=poly_feat_names)

# Add label values
poly_features_train['TARGET'] = poly_target

# Correlation between features and target values
poly_corrs = poly_features_train.corr()['TARGET'].sort_values()
```

```python
print('Top 5 polynomial features:', poly_corrs.head(), sep='\n')
```

## Feature importances

```python
def plot_feat_importances(feature_importances):

    # Normalize feature importance
    feature_importances['importance_norm'] = feature_importances['importance'] / feature_importances['importance'].sum()
    
    # Sort by normalize importance 
    feature_importances = feature_importances.sort_values(by='importance_norm', ascending=False)
    
    # Plot feature importance
    fig, ax = plt.subplots(figsize=(10,8))

    ax.barh(feature_importances['feature'].head(15),
           feature_importances['importance_norm'].head(15).sort_values(ascending=False))

    #Inver y-axis so top feature is show at the top of the graph
    ax.invert_yaxis()

    plt.xlabel('Normalized importance'); plt.ylabel('Feature'); plt.title('Feature Importance (top 15)')
    plt.show()
```

## Light Gradient Boosted Machine

```python
from sklearn.model_selection import KFold
from sklearn.metrics import roc_auc_score
import lightgbm as lgb
```

```python
def lgb_model(train, test, encoding='ohe', n_folds=5):
    
    # Extract ids
    train_idx = train['SK_ID_CURR']
    test_idx = test['SK_ID_CURR']
    
    # Extract label column
    labels = train['TARGET']
    
    # Remove id and target columns
    train = train.drop(columns=['SK_ID_CURR', 'TARGET'])
    test = test.drop(columns=['SK_ID_CURR'])
    
    # One hot encoding
    if encoding == 'ohe':
        # Genereta one hot vectors
        train = pd.get_dummies(train)
        test = pd.get_dummies(test)
        
        # Align dataframes
        train, test = train.align(test, join='inner', axis=1)
        
        # No categorical indices
        cat_indices = 'auto'
    
    # Label encoding   
    elif encoding == 'le':
        
        # Create encoder
        le = LabelEncoder()
        
        # Store categorical indices
        cat_indices = []
        
        for i, col in enumerate(train.columns):
            if train[col].dtype == 'object':
            
                # Transform categorical features to integers
                train[col] = le.fit_transform(train[col])
                test[col] = le.transform(test[col])
            
                # Store indices
                cat_indices.append(i)
            
    else:
        raise ValueError('Must choose encoding ("ohe" or "le")')
        
    print('Training set shape:', train.shape)
    print('Testing set shape:', test.shape)
    
    # Feature names
    feature_names = list(train.columns)
    
    # Convert train, test sets to numpy arrays
    train = np.array(train)
    test = np.array(test)
    
    # K fold
    k_fold = KFold(n_splits=n_folds, shuffle=True, random_state=50)
    
    # Array for features importances
    feature_importance_vals = np.zeros(len(feature_names))
    
    # Array for predictions
    predictions = np.zeros(test.shape[0])
    
    # Array for out of fold predictions
    out_fold = np.zeros(train.shape[0])
    
    # Validation and training scores
    valid_scores = []
    train_scores = []
    
    # Iterate folds
    for train_indices, valid_indices in k_fold.split(train):
        
        # Training fold
        train_features, train_labels = train[train_indices], labels[train_indices]
        
        # Validation fold
        valid_features, valid_labels = train[valid_indices], labels[valid_indices]
        
        # LGBM model
        model = lgb.LGBMClassifier(n_estimators=10000, objective='binary', class_weight='balanced',
                                  learning_rate=0.05, reg_alpha=0.1, reg_lambda=0.1,
                                  subsample=0.8, n_jobs=-1, random_state=50)
        
        # Train model
        model.fit(train_features, train_labels, eval_metric='auc',
                 eval_set=[(valid_features, valid_labels), (train_features, train_labels)],
                 eval_names=['valid', 'train'], categorical_feature=cat_indices,
                 early_stopping_rounds=100, verbose=200)
            
        
        # Best iteration
        best_iteration = model.best_iteration_
        
        # Feature importances
        feature_importance_vals += model.feature_importances_ / k_fold.n_splits
        
        # Generate predictions
        predictions += model.predict_proba(test, num_iteration=best_iteration)[:, 1] / k_fold.n_splits
        
        # Out of fold predictions
        out_fold[valid_indices] = model.predict_proba(valid_features, num_iteration=best_iteration)[:, 1]
        
        # Best score
        valid_score = model.best_score_['valid']['auc']
        train_score = model.best_score_['train']['auc']
        
        valid_scores.append(valid_score)
        train_scores.append(train_score)
        
    # Create dataframe for scores submission
        
    submission = pd.DataFrame({
            'SK_ID_CURR': test_idx,
            'TARGET': predictions
        })
        
     # Feature importances dataframe
    feature_importances = pd.DataFrame({
            'feature': feature_names,
            'importance': feature_importance_vals
        })
        
    # Overall validation score
    valid_auc = roc_auc_score(labels, out_fold)
        
    # Add overall socres to metrics
    valid_scores.append(valid_auc)
    train_scores.append(np.mean(train_scores))
        
    # Validation scores dataframe
    fold_names = list(range(n_folds))
    fold_names.append('overall')
        
    metrics = pd.DataFrame({
            'fold': fold_names,
            'train': train_scores,
            'valid': valid_scores
        })
    
    return submission, feature_importances, metrics
        
```

```python
submission, feats_importances, metrics = lgb_model(app_train, app_test)
print(metrics)
```
